```python
import pandas as pd
```

```python
from Utility import string2list
def unpack_param(df):
    if 'param' in df.columns:
        new = df.param.apply(string2list)
        return df.merge(pd.DataFrame(list(new)), left_index=True, right_index=True)
    else:
        return df 
```

```python
import numpy as np
def get_pipesegment(dataset_name):
    if dataset_name!='INET_Filled':
        df = pd.read_csv('scigrid/{}/data/{}_PipeSegments.csv'.format(dataset_name,dataset_name),sep=';')
    else:
        df = pd.read_csv('scigrid/{}/data/{}_PipeSegments.csv'.format(dataset_name,'INET'),sep=';')
    df = unpack_param(df)
    df.max_cap_M_m3_per_d = df.max_cap_M_m3_per_d.apply(lambda x: np.nan if x=='None' else x)
    return df
```

```python
compare_list = ['LKD_Raw','IGGIN','IGGINL','IGGI','INET_Raw','IGG','INET_Filled','NO_Raw']
```

```python
compare_dict = {}
for name in compare_list:
    compare_dict[name] = get_pipesegment(name)
```

```python
IGG_name = set(compare_dict['IGG'].id)
IGGINL_name = set(compare_dict['IGGIN'].id)
```

```python
i=0
concat=[]
for name,df in compare_dict.items():
    report = compare_dict[name].max_cap_M_m3_per_d.describe().reset_index()
    report.columns = ['index',name]
    report[name] = report[name].round(3)
    if i==0:  
        report.columns = ['index_',name]
        concat.append(report)
    else:
        concat.append(report[name])
report = pd.concat(concat,axis=1)
report.iloc[:,[0,1,3,5,7,9,11,13]]
```
